```python
import pandas as pd
import numpy as np
import re, gensim, json
from pyvi import ViTokenizer
```

```python
# Read data
with open('./data/UIT-ViIC/uitviic_captions_train2017.json', encoding="utf8") as f:
    train_data = f.read()

train_data = json.loads(train_data)
```

```python
sentences = []
```

```python
 for i in range(len(train_data['annotations'])):
    caption = train_data['annotations'][i]['caption']
    print(caption)

    tokens = ViTokenizer.tokenize(caption.lower()).split()
    sentences.append(tokens)
```

```python
sentences
```

```python
model = gensim.models.Word2Vec(sentences, vector_size=200, window=5, min_count=2, workers=4)
```

```python
# Save model
model.save("./word2vec model/gensim_word2vec_200dim_onDatac.model")
```
